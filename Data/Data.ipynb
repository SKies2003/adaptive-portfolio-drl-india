Nifty50 Constituents data downloaded manually from `https://www.niftyindices.com/indices/equity/broad-based-indices/NIFTY-Next-50/NIFTY-50`<br>

```python
import pandas as pd
df = pd.read_csv(r"Data\ind_nifty50list.csv")
df.head()
```

```python
df['Symbol'] = df['Symbol'] + '.NS'

industry_abbr = {
    'Financial Services': 'FIN',
    'Metals & Mining': 'MET',
    'Services': 'SER',
    'Healthcare': 'HEA',
    'Consumer Durables': 'CND',
    'Automobile and Auto Components': 'AUT',
    'Capital Goods': 'CAP',
    'Oil Gas & Consumable Fuels': 'OIL',
    'Telecommunication': 'TEL',
    'Fast Moving Consumer Goods': 'FMCG',
    'Construction Materials': 'CON',
    'Information Technology': 'IT',
    'Consumer Services': 'CSR',
    'Construction': 'CST',
    'Power': 'POW'
}

df['Industry_Abbr'] = df['Industry'].map(industry_abbr)
df['Industry_Symbol_Abbr'] = df['Industry_Abbr'] + '_' + df['Symbol']
df.head()
```

```python
import yfinance as yf

tickers = df['Symbol'].to_list()
data = yf.download(tickers, start = "2014-04-01", end = "2025-03-31")["Close"]
data.index = pd.to_datetime(data.index).strftime('%Y-%m-%d')
data.head()
```

```python
symbol_abbr = pd.Series(df.Industry_Symbol_Abbr.values, index = df.Symbol).to_dict()
data.rename(columns = symbol_abbr, inplace = True)
sector_sorted_columns = sorted(data.columns, key=lambda x: x.split('_')[0])
data = data[sector_sorted_columns]
data.head()
```

```python
# Function to identify stocks with more than two consecutive NaNs
def remove_stocks_with_consecutive_nans(data, max_consecutive_nans=2):
    # Create a mask to track NaNs (excluding 'Date' column)
    nan_mask = data.iloc[:, 1:].isna()  # Exclude the 'Date' column

    # Initialize a list to keep track of columns to drop
    columns_to_drop = []

    # Loop through each column (stock) to check for consecutive NaNs
    for col in nan_mask.columns:  # Iterate over columns except 'Date'
        # Check for consecutive NaNs in the column
        consecutive_nans = nan_mask[col].astype(int).groupby((~nan_mask[col]).cumsum()).cumsum()

        # If the maximum count of consecutive NaNs is greater than the threshold, mark the column for dropping
        if consecutive_nans.max() > max_consecutive_nans:
            columns_to_drop.append(col)

    # Drop the identified columns from the original DataFrame (preserving the Date column)
    data = data.drop(columns=columns_to_drop)

    return data

# Remove stocks with more than two consecutive NaNs
data = remove_stocks_with_consecutive_nans(data)

data.to_csv(r"Data\Close_price.csv")
data.head()
```

```python
import numpy as np
log_returns = np.log(data / data.shift(1))
log_returns.head()
```

```python
log_returns = log_returns.dropna(axis=1, how='all')
log_returns = log_returns.drop(log_returns.index[0])
log_returns.to_csv(r"Data\Log_returns.csv")
log_returns.head()
```

India Vix data files combined together as needed and saved

```python
import pandas as pd
import os
import glob # Library to find files matching a pattern

# Define the directory containing the VIX files
vix_data_dir = r"Data\India Vix" # Use raw string for Windows paths

# Define the desired date range
start_date = "2014-04-01"
end_date = "2025-03-31" # Adjust if your actual end date is different

# Define the expected column headers (based on your example)
# Note: Adjust these if your actual headers differ slightly (e.g., spacing, capitalization)
column_names = ['Date', 'Open', 'High', 'Low', 'Close', 'Prev. Close', 'Change', '% Change']

# Find all CSV files in the directory
# Using glob is often more flexible than os.listdir + filtering
csv_files = glob.glob(os.path.join(vix_data_dir, "*.csv"))

# Check if any files were found
if not csv_files:
    print(f"Error: No CSV files found in the directory: {vix_data_dir}")
else:
    print(f"Found {len(csv_files)} CSV files to process.")

    # List to hold individual DataFrames
    df_list = []

    # Loop through each found CSV file
    for file_path in csv_files:
        try:
            # Read the CSV, skipping the header row (header=None assumes no header)
            # skiprows=1 is another option if you are certain every file has exactly one header row
            temp_df = pd.read_csv(file_path, header=None, skiprows=1)
            df_list.append(temp_df)
            # print(f"Successfully read: {os.path.basename(file_path)}")
        except pd.errors.EmptyDataError:
            print(f"Warning: Skipping empty file: {os.path.basename(file_path)}")
        except Exception as e:
            print(f"Error reading file {os.path.basename(file_path)}: {e}")

    # Concatenate all DataFrames in the list
    if df_list: # Proceed only if at least one file was read successfully
        combined_vix_df = pd.concat(df_list, ignore_index=True)

        # Assign the correct column names
        combined_vix_df.columns = column_names

        # --- Data Cleaning and Formatting ---

        # 1. Convert 'Date' column to datetime objects
        #    Use the correct format string based on your example '01-APR-2014'
        try:
            combined_vix_df['Date'] = pd.to_datetime(combined_vix_df['Date'], format='%d-%b-%Y')
        except ValueError as e:
            print(f"Error parsing dates. Check date format in CSVs. Example error: {e}")
            # Optional: Add code here to inspect rows causing errors if needed
            # print(combined_vix_df[pd.to_datetime(combined_vix_df['Date'], format='%d-%b-%Y', errors='coerce').isna()])


        # 2. Sort DataFrame by date
        combined_vix_df = combined_vix_df.sort_values(by='Date')

        # 3. Set 'Date' as the index
        combined_vix_df = combined_vix_df.set_index('Date')

        # 4. Filter by the desired date range
        combined_vix_df = combined_vix_df.loc[start_date:end_date]

        # 5. Convert numeric columns to numeric types (float)
        #    Handle potential non-numeric values (like '-') by coercing them to NaN
        numeric_cols = ['Open', 'High', 'Low', 'Close', 'Prev. Close', 'Change', '% Change']
        for col in numeric_cols:
            # Strip whitespace that might interfere with conversion
            if combined_vix_df[col].dtype == 'object': # Only process if it's not already numeric
                 combined_vix_df[col] = combined_vix_df[col].astype(str).str.strip()
            combined_vix_df[col] = pd.to_numeric(combined_vix_df[col], errors='coerce')

        # 6. Optional: Handle potential duplicate dates (if files overlap)
        #    Keep the first occurrence for each date
        combined_vix_df = combined_vix_df[~combined_vix_df.index.duplicated(keep='first')]


        # --- Display Results ---
        print("\n--- Combined and Processed India VIX Data ---")
        print(f"Data range: {combined_vix_df.index.min()} to {combined_vix_df.index.max()}")
        print(f"Total trading days: {len(combined_vix_df)}")
        print("\nHead:")
        print(combined_vix_df.head())
        print("\nTail:")
        print(combined_vix_df.tail())

        # --- Optional: Save the combined data ---
        output_path = r"Data\India_Vix.csv"
        combined_vix_df.to_csv(output_path)
        print(f"\nCombined VIX data saved to: {output_path}")

    else:
        print("No dataframes were created. Check for errors during file reading.")
```
